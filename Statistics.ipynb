### 1. Beta


```python

number_of_datasets=int(input('Number of datasets: '))
#statistical models is better if we many trials
trials=int(input('Number of trials: '))

data1=beta(0.3,0.6, size=number_of_datasets)
header=['Trial 1']

for i in range(1,trials):
    header.append(f'Trial {i+1}')
    data1=np.vstack([data1,beta(0.3,0.6, size=number_of_datasets)])
    
df=pd.DataFrame(data=data1.T, columns=header)
    
    
```

```python
df2 = df.mean(axis=0)
plt.subplot(2,1,1)
sn.histplot(df2, bins=50, kde=True,cumulative=True)
plt.title('CDF')
plt.show()

plt.subplot(2,1,2)
sn.histplot(df2, bins=50, kde=True,cumulative=False, fill=False, color='red')
plt.title('PDF')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sn.kdeplot(df2)
plt.show()
```
